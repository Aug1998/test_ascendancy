```python
import json
import pandas as pd
from collections import Counter
import pprint

with open("data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

people = data["data"]

rows = []

for person in people:
    rows.append({
        "name": person.get("full_name"),
        "title": person.get("current_title"),
        "company": person.get("current_company_name"),
        "location": person.get("current_location"),
        "linkedin": person.get("linkedin_url"),
    })

df = pd.DataFrame(rows)

len(people)
Counter(len(person) for person in people)
```

```python
people_with_17_keys = (person for person in people if len(person) == 17)

pprint.pprint(next(people_with_17_keys))
```
